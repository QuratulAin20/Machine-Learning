```python
X= df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(type (y))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import make_classification
```

```python
X_train,X_test ,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=25)
```

```python
L =  LogisticRegression(max_iter=500, solver='lbfgs')
L.fit(X_train,y_train)
y_pred = L.predict(X_test)
```

```python
print(L.coef_)
print(L.intercept_)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score,f1_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

precision = precision_score(y_test,y_pred , average='binary')
print(f"Precision: {precision}")

recall = recall_score(y_test,y_pred,average='binary')
print(f"recall:{recall}")

f1 = f1_score(y_test,y_pred,average='binary')
print(f"F-1:{f1}")

```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)

print(report)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Hyper parameter tunning usimg Grid CV and Random Search

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
model = LogisticRegression()
panelty = ['l1', 'l2', 'elasticnet'] 
c_value = [10,1,0.1,0.001]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
max_iter= [500]

```

```python
params = dict(penalty=panelty,C=c_value,solver=solver,max_iter=max_iter)
```

```python
grid = GridSearchCV(estimator=model ,param_grid=params , cv=cv , scoring='accuracy')
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred1 = grid.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred1)
print(f"Accuracy: {accuracy}")
```

```python
random = RandomizedSearchCV(estimator=model, param_distributions=params, cv=5, scoring='accuracy')
```

```python
random.fit(X_train,y_train)
```

```python
random.best_params_
```
