```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('/kaggle/input/multiple-linear-regression-data/multiLinearRawData.csv')
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
R = Ridge(alpha=10000)
L = Lasso(alpha=1)
E =  ElasticNet(alpha=1,l1_ratio=0.9)
```

```python
# linear regression

lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
print('r2-score:',r2_score(y_test,y_pred))
```

```python
# Ridge
R.fit(X_train,y_train)
ry_pred = R.predict(X_test)
print('r2-score:',r2_score(y_test,ry_pred))
```

```python
# Lasso
L.fit(X_train,y_train)
ly_pred = L.predict(X_test)
print('r2-score:',r2_score(y_test,ly_pred))
```

```python
# Elastic net with more ratio ofridge than lasso

E.fit(X_train,y_train)
Ey_pred = E.predict(X_test)
print('r2-score:',r2_score(y_test,Ey_pred))
```
